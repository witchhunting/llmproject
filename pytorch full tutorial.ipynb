## 텐서의 연산(Operation)

```python
import math

a = torch.rand(1, 2) * 2 -1
print(a)
print(torch.abs(a))
print(torch.ceil(a))
print(torch.floor(a))
print(torch.clamp(a, -0.5, 0.5))
```

```python
print(a)
print(torch.min(a))
print(torch.max(a))
print(torch.mean(a))
print(torch.std(a))
print(torch.prod(a))
print(torch.unique(torch.tensor([1, 2, 3, 1, 2, 2])))
```

```python
x = torch.rand(2, 2)
print(x)
print(x.max(dim=0))
print(x.max(dim=1))
```

```python
x = torch.rand(2, 2)
print(x)
print(x.min(dim=0))
print(x.min(dim=1))
```

```python
x = torch.rand(2,2)
print(x)
y = torch.rand(2,2)
print(y)
```

```python
print(torch.add(x, y))
```

```python
result = torch.empty(2, 4)
torch.add(x, y, out = result)
result
```

```python
print(x)
print(y)
y.add_(x)
print(y)
```

```python
print(x)
print(y)
print(torch.matmul(x, y))
z = torch.mm(x, y)
print(z)
```

## 텐서의 조작(Manipulations)

```python
x = torch.Tensor([[1, 2],
              [3, 4]])
print(x)
print(x[0, 0])
print(x[0, 1])
print(x[1, 0])
print(x[1, 1])

print(x[:, 0])
print(x[:, 1])

print(x[0, :])
print(x[1, :])
```

```python
x = torch.randn(4, 5)
print(x)
y = x.view(20)
print(y)
z = x.view(5, -1)
print(z)
```

```python
x = torch.randn(1)
print(x)
print(x.item())
print(x.dtype)
```

```python
tensor = torch.rand(1, 3, 3)
print(tensor)
print(tensor.shape)
```

```python
t = tensor.squeeze()
print(t)
print(t.shape)
```

```python
t = torch.rand(3, 3)
print(t)
print(t.shape)
```

```python
tensor = t.unsqueeze(dim=2)
```

```python
print(tensor.shape)
tensor
```

```python
x = torch.FloatTensor([1, 4])
print(x)
y = torch.FloatTensor([2, 5])
print(y)
z = torch.FloatTensor([3, 6])
print(z)
print(torch.stack([x, y, z]))
```

```python
a = torch.randn(1, 3, 3)
print(a)
b = torch.randn(1, 3, 3)
print(b)
c = torch.cat((a, b), dim=0)
print(c)
print(c.size())
```

```python
c = torch.cat((a, b), dim=1)
print(c)
print(c.size())
```

```python
tensor = torch.rand(3, 6)
print(tensor)

t1, t2 = torch.chunk(tensor, 2, dim=1)
print(t1)
print(t2)
# print(t3)
```
